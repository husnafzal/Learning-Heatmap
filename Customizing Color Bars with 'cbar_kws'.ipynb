# Mastering Seaborn Heatmaps: Customizing Color Bars with 'cbar_kws'

```python
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
data = {'Fruit': ['Apple', 'Banana', 'Apple', 'Banana', 'Apple', 'Orange'],
        'Color': ['Red', 'Yellow', 'Green', 'Yellow', 'Red', 'Orange']}
df = pd.DataFrame(data)

# Create a cross tabulation
crosstab = pd.crosstab(df['Fruit'], df['Color'])
```

Selecting to showcase Color bar

```python
sns.heatmap(crosstab, cbar=True)
```

Labeling and Changing Orientation of Color bar

```python
sns.heatmap(crosstab, cbar=True, cbar_kws={"orientation": "horizontal", "label": "Frequency of Color Association"})
```

Narowing the Colour bar

```python
sns.heatmap(crosstab, cbar=True, cbar_kws={"shrink": 0.5})
```

Formatting of Tick Labels

```python
sns.heatmap(crosstab, cbar=True, cbar_kws={"format": "%.2f"})
```

 Customize Tick's Parameters

```python
sns.heatmap(crosstab, cbar=True, cbar_kws={"ticks": [0,1,2]})
```

Colour bar with Custom Aspect Ratio

```python
sns.heatmap(crosstab, cbar=True, cbar_kws={"aspect": 10})
```

Using Combined Customizations of Color bar

```python
sns.heatmap(crosstab, cbar=True, cbar_kws={"orientation": "horizontal", "label": "Colour Fruit Association", "shrink": 0.5, "ticks": [0, 1,2], "format": "%.2f", "aspect": 5})
```
